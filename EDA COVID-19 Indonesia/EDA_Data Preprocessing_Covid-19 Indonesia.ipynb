```python
numeric_features_with_not_null=[]
for i in numeric_features:
    if data[i].isnull().sum()==0:
        numeric_features_with_not_null.append(i)
numeric_features_with_not_null,len(numeric_features_with_not_null)
```

```python
row=col=0
fig,ax=plt.subplots(5,4,figsize=(20,20))
for i,feature in enumerate(numeric_features_with_not_null):
    if i%4==0 and i>0:
        row+=1
        col=0
    data[feature].plot(kind="kde",ax=ax[row,col])
    ax[row,col].set_title(feature)
    col+=1
    plt.title(feature)
    plt.tight_layout()
plt.show()
```
